```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv(r"C:\Users\VANSHIKA  JAIN\Downloads\titanic.csv")
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
inp = df.drop(['Survived','Ticket','Cabin','Name'], axis=1)
tar = df['Survived']
```

```python
inp
```

```python
Sex_le=LabelEncoder()
Embarked_le=LabelEncoder()
```

```python
inp['Sex']=sex_le.fit_transform(inp['Sex'])
```

```python
inp
```

```python
inp['Embarked']=sex_le.fit_transform(inp['Embarked'])
```

```python
inp
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(inp,tar,test_size=0.2, random_state=2)
```

```python
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)*100
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm=confusion_matrix(y_test,pred)
cm
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(100,100))
plot_tree(model, filled=True, feature_names=inp.columns, class_names=['Not Survived', 'Survived'])
plt.show()
```
